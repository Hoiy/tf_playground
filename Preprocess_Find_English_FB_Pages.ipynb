```python
import pickle
with open('link_posts.pkl', 'rb') as f:
    link_posts = pickle.load(f)
```

```python
link_posts
```

```python
import TextPreprocessing.Language.detector as ld
link_posts['lang_name'] = link_posts['name'].astype(str).apply(ld.detect)
link_posts['lang_description'] = link_posts['description'].astype(str).apply(ld.detect)
link_posts['lang_message'] = link_posts['message'].astype(str).apply(ld.detect)

#%timeit pages['name'].apply(lambda x: ld.detect(x, 'langdetect'))
#%timeit pages['about'].astype(str).apply(lambda x: ld.detect(x))
#%timeit pages['about'].astype(str).apply(lambda x: ld.detect(x, 'langdetect'))
```

```python
import numpy as np
import pandas as pd
eng_description = link_posts['lang_description'] == 'en'
eng_message = link_posts['lang_message'] == 'en'

en_link_posts = link_posts[eng_description & eng_message]
                                 
```

```python
en_link_posts.sample(100)
```

```python
list(en_link_posts)
```

Tally the number of english link post for each page

```python
tally = en_link_posts.groupby(['page_id']).size()
print(tally.describe())
print(en_link_posts.shape)
```

Filter out pages with less than 2 english posts

```python
en_pages = tally[tally >= 2]
print(en_pages.describe())
print(en_pages.sample(10))
```

```python
en_pages_id = en_pages.index.astype(str).tolist()
import json
with open('en_pages.json', 'w') as outfile:
    json.dump(en_pages_id, outfile)
```
